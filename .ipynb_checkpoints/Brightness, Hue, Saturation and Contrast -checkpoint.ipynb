```python
import cv2
import numpy as np

img=cv2.imread('colors1.jpg')
mask=np.zeros(img.shape,np.uint8)



# we want to increase or decrease brightness in real time. We want a mask to do that. 
# the mask will take the image. we apply filters on the mask and the mask is then added 
# to the image to produce the required brightness and darkness

# this is not a good operation. So the brightness and darkness is not the pixel value in this case!

def brighten(mask,amount):
    # np array
#     print mask.shape
    if len(mask.shape)==3:
        print 'here'
        b1,g1,r1=mask.shape
        for a in range(int(b1)):
            for b in range(int(g1)):
                for c in range(int(r1)):
                    if(mask[a][b][c]+amount>=0 and mask[a][b][c]+amount<=255):
                        mask[a][b][c]+=amount
                    else:
                        pass
    else :
        b1,g1=mask.shape
        for a in range(b1):
            for b in range(g1):
                if(mask[a][b]+amount>=0 and mask[a][b]+amount<=255):
                    mask[a][b]+=amount
                else :
                    pass
                
    return mask

# mask is the entity to operate on
while True:
    cv2.imshow('Image',img+mask)
    k= cv2.waitKey(1) & 0xFF
    if k==ord('q') or k==ord('Q'):
        cv2.destroyAllWindows()
        break
    elif k==ord('b') or k==ord('B'):
        mask=brighten(mask,10)
    elif k==ord('n') or k==ord('N'):
        mask=brighten(mask,-10)
    
cv2.destroyAllWindows()

        
        
```
