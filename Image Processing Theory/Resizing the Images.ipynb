# Uses interpolation to compute these. The extra points are cleverly interpolated.

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
img=cv2.imread('image.jpg')
length, width = img.shape[:2]
res = cv2.resize(img,(2*width, 2*length), interpolation = cv2.INTER_CUBIC)
cv2.imshow('image',res)
cv2.imshow('Orignal',img)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import cv2
import numpy as np

# these are rotate operations in open cv. Just do one thing, find the Matrix, then let the warpAffine do its job, that is it!

### for example:

img = cv2.imread('image.jpg',0)
rows,cols = img.shape

M = np.float32([[1,0,100],[0,1,50]])
dst = cv2.warpAffine(img,M,(cols,rows))
cv2.imshow('img',dst)

M = cv2.getRotationMatrix2D((cols/2,rows/2),90,1)
dst1 = cv2.warpAffine(img,M,(cols,rows))
print(dst1)
cv2.imshow('img1',dst1)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# affine transform

img = cv2.imread('image.jpg')
rows,cols,ch = img.shape

pts1 = np.float32([[50,50],[200,50],[50,200]])
pts2 = np.float32([[10,100],[200,50],[100,250]])

M = cv2.getAffineTransform(pts1,pts2)

dst = cv2.warpAffine(img,M,(cols,rows))

plt.subplot(121),plt.imshow(img),plt.title('Input')
plt.subplot(122),plt.imshow(dst),plt.title('Output')
plt.show()
```

```python
img = cv2.imread('sudoku.jpg')
rows,cols,ch = img.shape

pts1 = np.float32([[20,20],[368,52],[28,387],[389,390]])
pts2 = np.float32([[0,0],[300,0],[0,300],[300,300]])

M = cv2.getPerspectiveTransform(pts1,pts2)

dst = cv2.warpPerspective(img,M,(300,300))

cv2.imshow('Transformed image',dst)
cv2.waitKey(0)
cv2.destroyAllWindows()
plt.subplot(121),plt.imshow(img),plt.title('Input')
plt.subplot(122),plt.imshow(dst),plt.title('Output')
plt.show()
```
